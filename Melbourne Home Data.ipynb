```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
csv_path = r'../Datasets/Melbourne Housing Snapshot/melb_data.csv'
```

```python
melbourne_data = pd.read_csv(csv_path)
```

```python
y = melbourne_data.Price
```

```python
melbourne_features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
X = melbourne_data[melbourne_features]
```

```python
X.describe()
```

```python
decision_tree_model = DecisionTreeRegressor(random_state = 1)
decision_tree_model.fit(X, y)
y_pred_head = decision_tree_model.predict(X.head())
y_pred_head
```

```python
melbourne_data['Price'].head()
```

```python
y_pred = decision_tree_model.predict(X)
```

```python
pd.Series(y_pred).compare(y)
```

```python
mean_absolute_error(y, y_pred)
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, random_state=0)
X_train.shape, X_val.shape, y_train.shape, y_val.shape
```

```python
model = DecisionTreeRegressor(random_state=1)
```

```python
model.fit(X_train, y_train)
```

```python
y_val_pred = model.predict(X_val)
```

```python
mean_absolute_error(y_val, y_val_pred)
```
